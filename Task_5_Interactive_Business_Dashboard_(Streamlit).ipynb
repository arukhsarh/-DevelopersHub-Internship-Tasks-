```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

st.set_page_config(page_title="Business Dashboard", layout="wide")
st.title("📊 DevelopersHub Business Analytics Dashboard")

data = {
    'Category': ['Technology', 'Furniture', 'Office Supplies', 'Technology', 'Furniture'],
    'Region': ['East', 'West', 'North', 'South', 'Central'],
    'Sales': [1200, 800, 500, 1500, 700],
    'Profit': [300, 100, 50, 400, 80]
}
df = pd.DataFrame(data)

st.sidebar.header("Filter Data")
region = st.sidebar.multiselect("Select Region", options=df['Region'].unique(), default=df['Region'].unique())

filtered_df = df[df['Region'].isin(region)]

col1, col2, col3 = st.columns(3)
col1.metric("Total Sales", f"${filtered_df['Sales'].sum()}")
col2.metric("Total Profit", f"${filtered_df['Profit'].sum()}")
col3.metric("Profit Margin", f"{(filtered_df['Profit'].sum()/filtered_df['Sales'].sum())*100:.1f}%")

fig = px.bar(filtered_df, x='Category', y='Sales', color='Category', title="Sales by Category")
st.plotly_chart(fig)

st.write("Top Customers and Details:")
st.dataframe(filtered_df)
```
